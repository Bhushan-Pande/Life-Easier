```python
from transformers import pipeline
import gradio as gr
import tensorflow as tf
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
from gramformer import Gramformer
gf = Gramformer(models=1, use_gpu=True) # 0 = detector, 1 = highlighter, 2 = corrector, 3 = all
```

```python
def correct_article(inp):
  correction=gf.correct(inp)
  return correction



```

```python
output_text = gr.outputs.Textbox()
gr.Interface(correct_article,"textbox", output_text, title="Correct your statements",
             description="Correcting them with Transformers").launch()
```

```python
!pip install flask-ngrok
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = TFGPT2LMHeadModel.from_pretrained("gpt2", pad_token_id=tokenizer.eos_token_id)
```

```python
from flask_ngrok import run_with_ngrok
```

```python
from flask import Flask
from flask import Flask
from flask_cors import CORS, cross_origin
```

```python
from flask import request
from flask import Flask, redirect, url_for, render_template, request, flash
from transformers import pipeline
import gradio as gr
import tensorflow as tf
!pip install pprint
from pprint import pprint
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
summarizer=pipeline('summarization')
app=Flask(__name__, template_folder='/content/drive/MyDrive/particles.js-master')
run_with_ngrok(app)
CORS(app)
@app.route('/', methods=['GET', 'POST'])
def home():
  maindata=request.data.decode('utf-8')
  print(maindata)
  correction=gf.correct(maindata)
  print(correction)
  return render_template('index.html', data=correction)
  # return "ok"


@app.route('/summarizer1', methods=['GET', 'POST'])
def summarizer1():
  maindata=request.data.decode('utf-8')
  pprint(maindata)
  summarize=summarizer(maindata)
  pprint(summarize)
  return "hello"

@app.route('/sentenceformation', methods=['GET', 'POST'])
def sentenceformation():
  maindata=request.data.decode('utf-8')
  input_ids = tokenizer.encode(maindata, return_tensors='tf')
  beam_output = model.generate(input_ids, max_length=100, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
  output = tokenizer.decode(beam_output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
  pprint(".".join(output.split(".")[:-1]) + ".")
  return "hello"



app.run()
```
